```python
from math import pi     # to define pi

def cylinder(radius,height):
    # function returns Area and Volume of a sphere with radius R
    R = radius
    H = height
    A = 2*pi*R**2 + 2*pi*R*H
    V = pi*R**2*H
    return A, V

area, volume = cylinder(3,8) # this splits the tuple in two variables
print(f"the area is {area}, the volume is {volume}")

ratio = area/volume     
print(f"the ratio of the area divided by the volume is {ratio}")

```

```python
def istriple(a, b, c):
    if a**2 + b**2 == c**2 and a - int(a) == 0 and b - int(b) == 0 and c - int(c) == 0:
        return True
    else:
        return False


print(istriple(3,4,5))
print(istriple(5,12,13.7))
print(istriple(3,2,7))
```

```python
from math import *    
import numpy as np

x = np.random.normal(3, 5, 100) # Generate a Gaussian distribution

def stdev(y):  # Make the definition
    mean = np.sum(y)/len(y)
    points = (y - mean)**2  # before you can sum the "point minus mean squared"
    s = sqrt((np.sum(points)/len(y)))
    return s

s = stdev(x)
print(f"the standard diviation with stdev is {s}")

scheck = np.std(x) # To check the standard diviation
print(f"check of standard diviation {scheck}")

def stdevonel(y):  # Make the definition as a oneliner
    sonel = sqrt((np.sum(((y - (np.sum(y)/len(y)))**2))/len(y)))
    return sonel

sonel = stdevonel(x)
print(f"the standard diviation with stdevonel (as a oneliner) is {sonel}")
```

```python
import numpy as np
from math import *
disvel =[18.49, 2094.75, 35.48, 2320.87, 35.48, 1224.9, 11.34, 872.13, 7.12,  432.2, 9.82,  751, 23.66, 1410, 44, 2700, 66.7,  5292, 52.4,  2550, 55, 5253, 25.6,  2419, 26.19,2158]
dis = disvel[0::2] # Seperate teh lists
vel = disvel[1::2]

x = dis
y = vel

def pearson(X, Y):
    x = np.array(X)
    y = np.array(Y)
    rxy = (len(x)*sum(x*y)-sum(x)*sum(y))/(sqrt(len(x)*sum(x**2)-(sum(x))**2)*(sqrt(len(x)*sum(y**2)-(sum(y))**2)))
    return rxy

print(pearson(x,y))


```

```python
import numpy as np
from math import *
disvel =[18.49, 2094.75, 35.48, 2320.87, 35.48, 1224.9, 11.34, 872.13, 7.12,  432.2, 9.82,  751, 23.66, 1410, 44, 2700, 66.7,  5292, 52.4,  2550, 55, 5253, 25.6,  2419, 26.19,2158]
dis = disvel[0::2] # Seperate teh lists
vel = disvel[1::2]

x = np.array(dis)
y = np.array(vel)

b = (sum(x*y)/sum(x**2))

plt.scatter(x, y, marker = "o")
plt.grid()
plt.ylabel("Velocity") 
plt.xlabel("Distance")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("chi-squared fitting")
 

hubb = np.arange(0,80)
plt.plot(hubb, hubb*67.66, linestyle='-', label=r"$v=H_0D$", color= 'purple')

selc = np.arange(0,80)
plt.plot(hubb, selc*b, label=r"$Y=bX$", linestyle='-', color = 'orange')

plt.legend()  
plt.show()
```
